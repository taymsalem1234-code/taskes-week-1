Day 2

```python
def calculate_statistics (data: List[float]) -> Dict[str, float]:
    mean = sum(data) / len(data)
    minimum = min(data)
    maximum = max(data)
    data_range = maximum - minimum
    return {"mean": mean,"min": minimum,"max": maximum,"range": data_range}
def normalize_data (data: List[float], method: str = "minmax") -> List[float]:
    minimum = min(data)
    maximum = max(data)
    return [(x - minimum) / (maximum - minimum) for x in data]
def remove_outliers(data, threshold):
    return [x for x in data if x <= threshold]
def train_test_split(data, ratio):
    index = int(len(data) * ratio)
    return data[:index], data[index:]
def encode_labels(labels):
    unique = list(set(labels))
    return [unique.index(label) for label in labels]
numbers = [10, 20, 30, 40, 50, 100]
categories = ["AI", "Data", "AI", "Python"]
train, test = train_test_split(numbers, 0.8)
print("1. Statistics:", calculate_statistics(numbers))
print("2. Normalized:", normalize_data(numbers))
print("3. No Outliers (threshold 60):", remove_outliers(numbers, 60))
print(f"4. Split - Train: {train}, Test: {test}")
print("5. Encoded Labels:", encode_labels(categories))
```

```python
students = {
    "S001": {
        "name": "Alice Chen",
        "courses": {
            "CS101": {"grade": 92, "credits": 3},
            "MATH201": {"grade": 88, "credits": 4},
            "AI301": {"grade": 95, "credits": 3},
        },
        "advisor": "Dr. Smith",
    },
    "S002": {
        "name": "Bob Lee",
        "courses": {
            "CS101": {"grade": 85, "credits": 3},
            "MATH201": {"grade": 90, "credits": 4},
        },
        "advisor": "Dr. Patel",
    },
}
alice_ai_grade = students["S001"]["courses"]["AI301"]["grade"]
bob_courses = students["S002"]["courses"].values()
cs101_students = [s["name"] for s in students.values() if "CS101" in s["courses"]]
all_grades = [c["grade"] for s in students.values() for c in s["courses"].values()]
def calculate_gpa(student_data):
    courses = student_data["courses"].values()
    return sum(c["grade"] * c["credits"] for c in courses) / sum(c["credits"] for c in courses)
top_student = max(students.values(), key=calculate_gpa)
print(alice_ai_grade)
print(sum(c["grade"] * c["credits"] for c in bob_courses) / sum(c["credits"] for c in bob_courses))
print(cs101_students)
print(sum(all_grades) / len(all_grades))
print(top_student['name'])
```
